# Use the mean formula
$$
I = \int \phi(x)p(x) \underbrace{dx_{[a,b]}}_{Uniform\ pdf}
\quad \Rightarrow \quad
I \approx \frac{\sum_{i=1}^N \phi(x_i) \hat{p}(x_i)}{\sum_{i=1}^N \hat{p}(x_i)}
$$
where $\hat{p}$ may be the non-normalized prob. and  $x \sim Uniform[a,b].

```python
@partial(jit, static_argnums=(0,))
def Integ0(N, subkey):
    xi = jax.random.uniform(subkey, (N,), minval=-3,maxval=3)
    Z_hat = jnp.sum(prob(xi))
    return jnp.sum(f(xi))/Z_hat
```

# Importance Sampling
In case $p(x)\geq0$, but we cannot draw directly from it, we can use an **auxiliary distribution** $q(x)$ from which we can draw some samples.
$$
I= \int \phi(x) p(x) dx = \int \phi(x) \underbrace{p(x)/q(x)}_{w(x)}\ \underbrace{q(x)dx}_{aux.\ pdf}
$$

Then we can use the mean method $w_i=p(x_i)/q(x_i)$
$$
I = \frac{\sum_i \phi(x_i)w_i}{\sum_i w_i}
$$


Notice, at least the support of $q(x)$ must cover the $p(x)$ one, and we do not need to have $q$ normalized. 

Two $q(x)$ are considered below.

```python
scale_0 = 3
def qapp0(x):
    return 90*jax.scipy.stats.norm.pdf(x,scale=scale_0)
```

```python
scale_1 = 0.5
mean_1  = 1.5
def qapp(x):
    return 20*(jax.scipy.stats.norm.pdf(x,scale=scale_1)+\
            jax.scipy.stats.norm.pdf(x-mean_1,scale=scale_1))

```

```python
#Use sampling from the large single gaussian pdf
@partial(jit, static_argnums=1)
def importance_sampling_0(key,N):
    key, subkey = jax.random.split(key)
    x_Qrnd=jax.random.normal(subkey, (N,))*scale_0
    w = prob(x_Qrnd)/qapp0(x_Qrnd)
    norm = jnp.sum(w)
    phi_samples=phi(x_Qrnd)
    integ = jnp.dot(w/norm,phi_samples)
    return integ,w, key

#Use sampling from the two gaussians pdf
@partial(jit, static_argnums=1)
def importance_sampling(key,N):
    key, subkey = jax.random.split(key)
    xs1= jax.random.normal(subkey, (N//2,))*scale_1
    xs2= jax.random.normal(subkey, (N//2,))*scale_1 + mean_1
    x_Qrnd=jnp.concatenate([xs1,xs2])
    w = prob(x_Qrnd)/qapp(x_Qrnd)
    norm = jnp.sum(w)
    phi_samples=phi(x_Qrnd)
    integ = jnp.dot(w/norm,phi_samples)
    return integ,w, key
```

Better is the approximation q(x) of p(x), the better the integral is computed, i.e. the sampling of p(x). This is in fact reflected on the distribution of weights which at best should be concentrated around 1. But this becomes very unlikely in high dimension. See the following ex.

# Dimension d>1

Let be the uniform law in a ball of radius R
in dimension d:
$$
p(x) = \frac{1}{ V_d (R) } \begin{cases}
1 & if\ \sqrt{\sum_{i=1}^d x_i^2} = \|x\| = r < R \\
0 & if\ not
\end{cases}
$$
we want to calculate by *importance samping* the following integral
$$
I(d) = \int r^2 p(x) dx^d = R^2 \frac{d}{d+1}
$$
which is the 2nd moment of the p(x) pdf. Remember that we do not need to know the normalisation $V_d(R)$.

```python
@jit
def prob_nD(x,R=1.):
    r  = jnp.linalg.norm(x,axis=1)
    return jnp.where(r<R,1.,0.)
```

As auxiliary distribution $q$ we use the normal distribution centered in d dimensions:
$$
q(s) = \mathcal{N}_d(0,\sigma^2) = \frac{1}{(2\pi\sigma^2)^{d/2}}e^{-\frac{x^2}{2\sigma^2}}
$$

```python
@jit
def qapp_nD(x,s=1):
    dim = x.shape[1]
    return jax.scipy.stats.multivariate_normal.pdf(x,np.zeros(dim), s**2*np.identity(dim))
```

Where are located the points drawn according to a Gaussian distribution in d dimension?
$$
\langle r \rangle = \sigma d^{1/2}; \qquad \sigma_r = \frac{\sigma}{\sqrt{2}}
$$

```python
key = jax.random.PRNGKey(0)
N=10_000
dims = jnp.array([10,20,50,100,200,500,1000])
s=1
info1 = []
info2 = []
for dim in dims:
    key, subkey = jax.random.split(key)
    data = jax.random.multivariate_normal(subkey,
                                        jnp.zeros(dim),
                                        s**2*jnp.identity(dim),
                                        (N,))
    r = jnp.linalg.norm(data,axis=1)
    info1.append(jnp.mean(r))
    info2.append(jnp.std(r))
```

```python
fig, ax = plt.subplots()
axes = [ax, ax.twinx()]
axes[0].scatter(dims,info1,c='b',label="<r>");
xi = np.arange(1,1000,0.1)
axes[0].plot(xi,np.sqrt(xi),c='b',ls="--")
axes[0].set_xlabel("d")
axes[0].set_ylabel("$<r>$")
axes[1].scatter(dims,info2,c='r',label="$\sigma_r$");
axes[1].axhline(y=1/np.sqrt(2), color='r', linestyle='--')
axes[1].set_ylim([0.6,0.8])
axes[1].set_ylabel("$\sigma_r$")
plt.title(r"$\langle r \rangle$ et $\sigma_r$, loi $\mathcal{N}_d(0,1)$");
axes[0].legend(loc='upper left');
axes[1].legend(loc="lower right");
```

So in d dimensions, the points generated by a centered Gaussian, contrary to intuition, are concentrated not around the origin but on a shell of radius $\sigma \sqrt{d}$
whose thickness is a constant $\sigma_r$. Therefore, in order for the weights $p(x_i)/q(x_i)$ are not zero or infinite it is necessary that $\langle r \rangle \leq R$ that is to say $\sigma \leq R/\sqrt{d}$.

```python
@jit
def phi_nD(x):
    r  = jnp.linalg.norm(x,axis=1)
    return r*r
```

```python
@partial(jit, static_argnums=(1,2))
def importance_sampling_nD(key,N,dim=2,R=1,s=1):
    key, subkey = jax.random.split(key)
    x_Qrnd = jax.random.multivariate_normal(subkey,
                                            jnp.zeros(dim),
                                            s**2*jnp.identity(dim),
                                            (N,))
    
    w = prob_nD(x_Qrnd,R)/qapp_nD(x_Qrnd,s)
    norm = jnp.sum(w)
    w /= norm
    phi_samples=phi_nD(x_Qrnd)
    integ = jnp.dot(w,phi_samples)
    return key,integ,w
```

# Metropolis-Hastings method

The method dates back to the 1950s-70s. In the Importance Sampling $q(x)$
is chosen once and for all. M-H will generate samples constituting a **Markov chain**, where **$q(x)$ will be adapted to each titration.**

To sample a sample $x_i$ at step $i$:


1.   first sample $\hat{x}\sim q(x; x_{i-1})$; for instance $q$ can be a gaussian distribution of mean $x_{i-1}$
2.   then one defines the following ratio
$$
r = \frac{\tilde{p}(\hat{x})}{\tilde{p}(x_{i-1})} \times \frac{q(x_{i-1}; \hat{x})}{q(\hat{x}; x_{i-1})}  \qquad (Metropolis-Hastings)
$$
3. finaly, to get $x_i$:
$$
u \sim U(0,1), \qquad 
x_i = 
\begin{cases}
\hat{x} & r>u \\
x_{i-1} & overwise
\end{cases}
$$

The initialisation is performed by sampling $x_0$. The pdf $q(x;y)$ is the *transition probabilit*. If $q(x;y) = q(y;x)$ (rversibility) then the ratio $r$ is simpler
$$
r =  \frac{\tilde{p}(\hat{x})}{\tilde{p}(x_{i-1})} \qquad (Metropolis)
$$ 

## Simple cas en 1D, multi-chaines (Numpy)

```python
class MixtureModel(scs.rv_continuous):
    def __init__(self, submodels, weights, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.submodels = submodels
        weights   = np.array(weights)
        norm = np.sum(weights)
        self.weights = weights/norm
        self.num_distr = len(submodels)
        #print(self.num_distr, self.weights,self.submodels)
        

    def _pdf(self, x):
        pdf = 0.
        for i, submodel in enumerate(self.submodels):
            pdf += self.weights[i]*submodel.pdf(x)
            
        return pdf

    # Random generation to compare with Metropolis sampling
    def rvs(self, sample_size):
        random_idx = np.random.choice(np.arange(self.num_distr), size=(sample_size,), p=self.weights)
        submodel_samples = [submodel.rvs(size=sample_size) for submodel in self.submodels]
        print(np.array(submodel_samples).shape)
        rvs = np.choose(random_idx, submodel_samples)
        return rvs
```

The idea is to use the M-H method to generate samples according to $p(x)$
(mixture of gaussian) is to compute the average of $\phi(x)$.

```python
def MHSampling_v2p(proba,Ns=100_000,Nchain=10,scale=0.1, forget=25_000):
    # Ns nbre de samples per chain
    samples = []
    # forget a fixed number of samples for each chain
    Ns = Ns + forget
    current = scs.uniform.rvs(loc=-3,scale=6, size=(Nchain,))
    for i in range(Ns):
        samples.append(current.copy())
        #proposal distrib. uniforme in [current-scale,current+scale]
        proposal = scs.uniform.rvs(loc=current-scale,scale=2*scale, size=(Nchain,))
        # proposal distrib is symetric: q(x_c,x_p)=q(x_c-x_p)=q(x_p,x_c)
        r = proba(proposal)/proba(current)  
        u = scs.uniform.rvs(size=(Nchain,))
        mask = r>u #accept proposal
        current[mask] = proposal[mask]

    #return all chains with the first "forget" samples rejected
    samples = np.array(samples).T
    return samples[:,forget:]
```

Below : a version with logprob and implemented in Jax. Here just to see that JAX/JIT is faster but does not change the M-H problem.

```python
# One possibility to use jit on function in the class is to use a off-class function 
# this is due to the self arg
# There are alternatives w/o any class
@jit
def _logpdf(x, loc, scale, weights):
  log_probs = jax.scipy.stats.norm.logpdf(x,loc=loc, scale=scale)
  return jax.scipy.special.logsumexp(jnp.log(weights) + log_probs, axis=0)

class MixtureModel_jax():
    def __init__(self, locs, scales, weights, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.loc = jnp.array([locs]).T
        self.scale = jnp.array([scales]).T
        self.weights = jnp.array([weights]).T
        norm = jnp.sum(self.weights)
        self.weights = self.weights/norm

        self.num_distr = len(locs)

    def pdf(self, x):
        probs = jax.scipy.stats.norm.pdf(x,loc=self.loc, scale=self.scale)
        return jnp.dot(self.weights.T,probs).squeeze()
    
    def logpdf(self, x):
        return _logpdf(x, self.loc, self.scale, self.weights)
```

```python
mixture_gaussian_model = MixtureModel_jax([0,1.5],[0.5,0.1],[8,2])
```

```python
@partial(jax.jit, static_argnums=(1,))
def jax_metropolis_kernel(rng_key, logpdf, position, log_prob):
    key, subkey = jax.random.split(rng_key)
    """Moves the chain by one step using the Random Walk Metropolis algorithm.
  
    Attributes
    ----------
    rng_key: jax.random.PRNGKey
      Key for the pseudo random number generator.
    logpdf: function
      Returns the log-probability of the model given a position.
    position: np.ndarray, shape (n_dims,)
      The starting position.
    log_prob: float
      The log probability at the starting position.

    Returns
    -------
    Tuple
        The next positions of the chains along with their log probability.
    """
    move_proposals = jax.random.normal(key, shape=position.shape) * 0.1
    proposal = position + move_proposals
    proposal_log_prob = logpdf(proposal)

    log_uniform = jnp.log(jax.random.uniform(subkey))
    do_accept = log_uniform < proposal_log_prob - log_prob

    position = jnp.where(do_accept, proposal, position)
    log_prob = jnp.where(do_accept, proposal_log_prob, log_prob)
    return position, log_prob

@partial(jax.jit, static_argnums=(1, 2))
def jax_metropolis_sampler(rng_key, n_samples, logpdf, initial_position):
    """Generate samples using the Random Walk Metropolis algorithm.
    Attributes
    ----------
    rng_key: jax.random.PRNGKey
        Key for the pseudo random number generator.
    n_samples: int
        Number of samples to generate per chain.
    logpdf: function
      Returns the log-probability of the model given a position.
    inital_position: np.ndarray (n_dims, n_chains)
      The starting position.
    Returns
    -------
    (n_samples, n_dim)
    """
    def mh_update(i, state):
        key, positions, log_prob = state
        _, key = jax.random.split(key)
        new_position, new_log_prob = jax_metropolis_kernel(key, 
                                                           logpdf, 
                                                           positions[i-1], 
                                                           log_prob)
        positions=positions.at[i].set(new_position)
        return (key, positions, new_log_prob)


    logp = logpdf(initial_position)
    # all positions structure should be set before lax.fori_loop
    
    all_positions = jnp.zeros((n_samples,initial_position.shape[0]))  # 1 chain for each vmap call
    
    initial_state = (rng_key,all_positions, logp)
    rng_key, all_positions, log_prob = jax.lax.fori_loop(1, n_samples, 
                                                 mh_update, 
                                                 initial_state)
    
    
    return all_positions
```

```python
@jit
def jax_my_logpdf(par,xi,yi):
    # priors: a=par[0], b=par[1], sigma=par[2]
    logpdf_a = jax.scipy.stats.norm.logpdf(x=par[0],loc=0.5,scale=1.)
    logpdf_b = jax.scipy.stats.norm.logpdf(x=par[1],loc=2.5,scale=1.)
    logpdf_s = jax.scipy.stats.gamma.logpdf(x=par[2],a=3,scale=1.)

    val = xi*par[1]+par[0]
    tmp = jax.scipy.stats.norm.logpdf(x=val,loc=yi,scale=par[2])    
    log_likeh= jnp.sum(tmp)
    
    return log_likeh + logpdf_a + logpdf_b + logpdf_s
```

# Hamiltonian Monte Carlo

HMC in 180sec: we have a target probability $p_x(x)$ that we  would like to sample. Let us right it as in Statistical Physics
$$
p_c(x) = \frac{1}{Z} e^{-U(x)} \qquad U(x) = -\log p_c(x) = -\log[\pi(x)\mathcal{L}(x|\mathcal{D})]
$$
$\pi(x)$ is a prior and $\mathcal{L}(x|\mathcal{D}))$ the likelihood with $\mathcal{D}$ the dataset.

Then the HMC defines a auxiliary varaible $p$ and
$$
H(x,p) = U(x) + E_k(p)
$$
with $E_k(p)$ a pseudo kinetic energy with possible mass matrix ($p^T M^{-1} p$). The Hamiltonian dynamical equations read
$$
\dot{x} = \frac{\partial H}{\partial p} = \frac{\partial E_k}{\partial p} = M^{-1} p  \qquad  \dot{p} = - \frac{\partial H}{\partial x} = - \frac{\partial U}{\partial x} = \frac{\partial \log p_c(x)}{\partial x}
$$ 

Let us then define the following algorithm


1.   Sample a momentum from $p_i \sim \mathcal{N}(0,1)$, and set $p_{new}=p_i$ and $x_{new}=x_i$
2.   Iteration of the dynamic eqs with the leap-frog algo of $n_{steps}$ and step size $$ with continuous update of ($p_{new}, x_{new}$):
$$
\begin{cases}
p_{new} &=  p_{new} -  \frac{1}{2} \frac{\partial U}{\partial x}|_{x=x_{new}} \\
x_{new} &= x_{new} +  M^{-1} p_{new} \\
p_{new} &= p_{new} -  \frac{1}{2} \frac{\partial U}{\partial x}|_{x=x_{new}}
\end{cases}
$$
3. then we possibly reverse the moment $p_{new} = - p_{new}$ (not necessay in practice)
4. finally, we proceed as for the Metropolis algorithm according to the acceptance probability of the new state $p_{new}, x_{new}$): 
$$
p_{acc}  = \min\left\{1, r= \frac{\hat{P}(x_{new},p_{new})}{\hat{P}(x_i,p_i)} \right\}
$$
with 
$$
\hat{P}(x,p) = e^{-H(x,p)}
$$
So one shoot a number $u$ uniform dist. in $[0,1]$, then if $u < r$ accept $x_{new}$ for $x_{i+1}$ overwise keep $x_i$. In this algorithm the momentum $p$ in drawn at each iteration w/o memory.



## A code from scratch (Numpy)

```python
def normal(x,mu,sigma):
    return np.exp(-neg_log_prob(x,mu,sigma))

def neg_log_prob(x,mu,sigma):
#    print(x,mu,sigma)
    tmp= 0.5*((x-mu)/sigma)**2 + 0.5*np.log(2*np.pi*sigma**2)
    return tmp

def gradV(x,mu,sigma):
    return (x-mu)/sigma**2

def HMC(mu=0.0,sigma=1.0,n_steps=10,step_size=0.25,initial_position=0.0,n_spl=10,dump=False):
    # setup
    samples = np.zeros(n_spl)
    samples[0] = initial_position
    momentum_dist = scs.norm(scale=1.0) # gaussian distribution for momentum proposal
    n_accept = 1
    # generate samples
    for i in range(1,n_spl):
        q0 = samples[i-1]
        q1 = q0
        p0 = momentum_dist.rvs()        
        p1 = p0 
        gradV0 = gradV(q0,mu,sigma) #(q0-mu)/(sigma**2) # explicit gradient

        
        # leapfrog integration begin
        for s in range(n_steps): 
            p1 -= step_size*gradV0/2 # as potential energy increases, kinetic energy decreases, half-step
            q1 += step_size*p1     # position increases as function of momentum 
            gradV1 = gradV(q1,mu,sigma) #(q1-mu)/(sigma**2) # update gradient
            if dump:
                print(f"[{i}/{s}] gradV0:{gradV0}, gradV1:{gradV1}")
            p1 -= step_size*gradV1/2 # second half-step "leapfrog" update to momentum    
        # leapfrog integration end        
        #### p1 = -1*p1 #flip momentum for reversibility unnecessary in practice as kinetic term is quadratic in p    

        
        #metropolis-hastings acceptance (negative log version)
        q0_nlp = neg_log_prob(x=q0,mu=mu,sigma=sigma)
        q1_nlp = neg_log_prob(x=q1,mu=mu,sigma=sigma)        

        p0_nlp = neg_log_prob(x=p0,mu=0.,sigma=1.)
        p1_nlp = neg_log_prob(x=p1,mu=0.,sigma=1.)
        
        target = q0_nlp - q1_nlp # P(q1)/P(q0)
        adjustment = p1_nlp - p0_nlp # P(p0)/P(p1)
        r = target + adjustment # [P(q1)*P(p0)]/[P(q0)*P(p1)] 
        
        u = np.log(scs.uniform.rvs())
            
        if dump:
            print(f"[{i}] q0:{q0},q1:{q1}")

        
        if u < r:
            samples[i]=q1
            n_accept += 1
        else:
            samples[i]=q0
    
    return samples, n_accept/n_spl
```

```python
mu = 0.
sigma = 1.
Ns = 10_000
samples, acc_rate = HMC(n_steps=9,step_size=0.23, n_spl=Ns, 
                      initial_position=scs.uniform.rvs(loc=-3,scale=6),
                      mu=mu,sigma=sigma,dump=False) 
print('accept rate:',acc_rate)
```

# Takeaway :
- in 1D we can calculate integrals with quadratures (Simpson, Trapezium, Gauss, Clenshaw-Curtis, etc.) but in large dimensions, we can only proceed with Monte Carlo sampling methods. 

- First we have seen the Importance Sampling using an auxilary pdf $q(x)$
which approximates $p(x)$. But this only works well if $q$ is a "good" approximation $p$. Moreover, this is more and more required as the dimension ncreases.

- By adapting $q(x)$ along the sampling, we end up with the Metropolis-Hastings method. But this one, although often used, has a sampling efficiency of 1% or even lower, and this occurs even on not too pathological cases in 1D. This is due to the random walk character of the method.

- A last promising method is the one based on Hamilton's differential equations (Hamiltionian Monte carlo, aka HMC). It increases the sampling efficiency by a factor of 10, but the parameters of the leapfrog algorithm must be chosen carefully. This is why, No-U-Turns algo for instance has been developped (subject of an other nb)
